```python
import pandas as pd

def analyze_words(words, nlp, lang="fr"):
    data = []
    for word in words:
        doc = nlp(word['mot'])
        for token in doc:
            # Comparaison modèle vs JSON
            lemme_accuracy = 1 if token.lemma_ == word["lemme"] else 0
            prefix_accuracy = 1 if token.prefix_ == word["prefixe"] else 0
            suffix_accuracy = 1 if token.suffix_ == word["suffixe"] else 0

            data.append({
                "Langue": lang,
                "Mot original": token.text,
                "Lemme (modèle)": token.lemma_,
                "Lemme (référence)": word["lemme"],
                "Préfixe (modèle)": token.prefix_,
                "Préfixe (référence)": word["prefixe"],
                "Suffixe (modèle)": token.suffix_,
                "Suffixe (référence)": word["suffixe"],
                "POS": token.pos_,
                "Morphologie": token.morph,
                "Lemme Accuracy": lemme_accuracy,
                "Prefix Accuracy": prefix_accuracy,
                "Suffix Accuracy": suffix_accuracy
            })
    
    return pd.DataFrame(data)

# Analyse des mots français et anglais
df_fr = analyze_words(words_data["francais"], nlp_fr, "fr")
df_en = analyze_words(words_data["anglais"], nlp_en, "en")
```

```python
# Calcul des précisions
def print_accuracy(lang, df):
    lemme_acc = df["Lemme Accuracy"].mean()
    prefix_acc = df["Prefix Accuracy"].mean()
    suffix_acc = df["Suffix Accuracy"].mean()

    print(f"--- Précision pour la langue: {lang.upper()} ---")
    print(f"Lemme accuracy     : {lemme_acc * 100:.2f}%")
    print(f"Préfixe accuracy   : {prefix_acc * 100:.2f}%")
    print(f"Suffixe accuracy   : {suffix_acc * 100:.2f}%\n")

print_accuracy("fr", df_fr)
print_accuracy("en", df_en)
```

```python
import nltk.tokenize as to #type:ignore
from nltk.stem.snowball import SnowballStemmer #type:ignore
# Charger un modèle pré-entraîné en français

import re

def __sub_morph(text=None):
        """
        sert a decouper les mot pour le morph
        """

        regex = r'[-]|[_]|[^\wÀ-ÿ\'\’\s\-"]' #\w équivalent à la classe [a-zA-Z0-9_]. \s équivalent à la classe [ \t\n\r\f\v].
        #la les mots comme pensa-t-elle vaut 1  quelqu'un aussi vaut 1 pour que il vale deux faut raouter \' et \’
        text=re.sub(regex,"",text) #re.sub(pattern, repl, string, count=0, flags=0)
        regex = r"[-]|[_]" #ici pensa t'elle vaudrais 3
        words=re.sub(regex," ",text)
        return to.word_tokenize(words)

def morphem(text=None):
        """
        retourne un dictionnaire pour chaque mot du text avec trois valeur prefixe infixe et suffixe qui sont de bool
        """
        words=__sub_morph(text)
        words=set(words)
        word_dict={}
        stemmer = SnowballStemmer("french")
        for word in words:
            word_dict[word]={
                "prefixe":"",
                "infixe":"",
                "suffixe":""
            }
            if word and word[0] in prefixes:

                for prefix in prefixes[word[0]]:
                    if word.startswith(prefix) and len(word)!=len(prefix):
                        word_dict[word]["prefixe"]=prefix
                        break

            stem=stemmer.stem(word)
            word_suffixe= word.replace(stem, "")
            while len(word_suffixe) > 0 and word_dict[word]["suffixe"] == False:
                """
                On cherche les suffixes, parfois les suffixes ne sont pas directs comme dans 'constitutionnellement'.
                Ici, la racine donnerait 'constitutionnel', et le suffixe serait 'ment', mais parfois, on doit retirer
                des lettres pour trouver le bon suffixe. Par exemple, dans 'constitutionnellement', le suffixe est 'ment'.
                """

                # Vérifie si le suffixe actuel existe dans les suffixes et si le mot se termine par ce suffixe
                if word_suffixe[0] in suffixes:
                    # Trouver les suffixes correspondants
                    for suf in suffixes[word_suffixe[0]]:
                        if word.endswith(suf) and len(suf)>1:  # Si le mot se termine par le suffixe
                            word_dict[word]["suffixe"] = suf  # Assigner le suffixe trouvé
                            break  # # Sortir de la boucle dès qu'un suffixe est trouvé

                # Si le suffixe n'a pas été trouvé, passer au suffixe suivant
                word_suffixe = word_suffixe[1:]

            for cle,liste in suffixes.items():
                if word_dict[word]["infixe"]==True:
                    break
                for suf in liste:
                    #un infixe peut pas etre egale a la taille du mot il peut pas mesurer 1 et un le mot moin infixe peut pas donner moin de 1
                    if stem.endswith(suf) and len(stem)!=len(suf) and (len(stem)-len(suf))>1 and len(suf)>1:
                        word_dict[word]["infixe"]=True
                        break


        return word_dict
```
